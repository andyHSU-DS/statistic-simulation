```python
import pandas as pd
```

```python
data=pd.read_csv(r'/Users/andyhsu/Desktop/statistic/ANOVA/consumption_data.csv',encoding='big5')
```

```python
data
```

```python
alist = data[data ['廣告'] == '廣告1']['消費金額'].tolist()
blist = data[data ['廣告'] == '廣告2']['消費金額'].tolist()
clist = data[data ['廣告'] == '廣告3']['消費金額'].tolist()
dlist = data[data ['地區'] == '北部']['消費金額'].tolist()
elist = data[data ['地區'] == '中部']['消費金額'].tolist()
flist = data[data ['地區'] == '南部']['消費金額'].tolist()
```

# 事前檢定

```python
import scipy.stats as st
a=1
for x in [alist,blist,clist,dlist,elist,flist]:
    print(str(a),':',st.shapiro(x)[1])
    a+=1
```

# 同質性檢定

```python
print(st.levene(alist, blist, clist, center='mean'))
print(st.levene(dlist, elist, flist, center='mean'))#因為拆分依據不同，所以不能一起比
```

# 多因子分析

```python
data
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('消費金額 ~ 廣告 + 地區 + 廣告:地區', data=data).fit()
```

```python
model.summary()
```

```python
Results = sm.stats.anova_lm(model)
```
